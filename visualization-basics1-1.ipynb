```python
#Line Plots: to understand the trends and patterns in data
z = np.linspace(0, 2, 100) # no. of samples to generate

#Step 1: Declaring the figure and the axis
# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots()  # Create a figure and an axes.

#Step 2: start feeding data; .plot method generates line plot
ax.plot(z, z, label='linear')  # Plot some data on the axes. #plot 1

ax.plot(z, z**2, label='quadratic')  # Plot more data on the axes... #plot 2

ax.plot(z, z**3, label='cubic')  # ... and some more.

#step 3: Naming your x and y axis
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.

ax.set_title("Simple Plot")  # Add a title to the axes.

ax.legend()  # Add a legend.

plt.show()
```
